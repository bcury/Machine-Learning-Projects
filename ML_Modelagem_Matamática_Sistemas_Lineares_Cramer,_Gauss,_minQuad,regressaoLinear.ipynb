Função regra de cramer para sistema com 3 incógnitas

```python
matriz
```

```python
np.array([matriz[:,0],matriz[:,1],matriz[:,2]]) #assim invertemos coluna com linha
```

```python
np.array([matriz[:,0],matriz[:,1],matriz[:,2]]).T # agora voltamos ela ao normal, usando a transposta
```

```python
def cramer3 (A,v):
  determinante = sp.linalg.det(np.array([A[:,0],A[:,1],A[:,2]]).T)
  if determinante !=0:
    detx = sp.linalg.det(np.array([v[:],A[:,1],A[:,2]]).T)
    dety = sp.linalg.det(np.array([A[:,0],v[:],A[:,2]]).T)
    detz = sp.linalg.det(np.array([A[:,0],A[:,1],v[:]]).T)
    x = detx/determinante
    y = dety/determinante
    z = detz/determinante
    print(x)
    print(y)
    print(z)
  else:
    print('coef nulo')
```

```python
cramer3(matriz, vetor)
```

criação da matriz ds constantes

```python
b = nota.drop(columns=['inicio_estudo','tempo_estudo_dia'])
b
```

```python
b = b.values
print(b)
```

```python
#estimativa min quadrados

```

```python
At = np.array(A).T
At
```

```python
#mult da At pela A
AtA = np.dot(At,A)
AtA
```

```python
#inversa da AtA
inversa = np.linalg.inv(AtA)
inversa
```

```python
Atb = np.dot(At, b)
Atb
```

```python
x = np.dot(inversa, Atb)
x
```

ou

```python
x = np.dot(np.dot(np.linalg.inv(np.dot(At,A)),At),b)
x
```

```python
resultado = nota
resultado
```

```python
resultado['previsao'] = resultado['inicio_estudo']*x[0] + resultado['tempo_estudo_dia']*x[1]
resultado
```

erro médio absoluto

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(resultado.nota, resultado.previsao)
```

```python
resultado['erro_abs'] = abs(resultado.nota - resultado.previsao)
```
