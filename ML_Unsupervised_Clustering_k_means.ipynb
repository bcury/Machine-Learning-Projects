```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.cluster import KMeans
```

```python
x = [20, 27, 21, 37, 46, 53, 55, 47, 52, 32, 39, 41, 39, 48, 48] # base de idades
y = [1000, 1200, 2900, 1850, 900, 950, 2000, 2100, 3000, 5900, 4100, 5100, 7000, 5000, 6500] #base de salarios
```

```python
grafico = px.scatter(x = x, y = y)
grafico.show()
```

```python
base_salario = np.array([[20,1000],[27,1200],[21,2900],[37,1850],[46,900],[53,950],[55,2000],[47,2100],[52,3000],[32,5900],[39,4100],[41,5100],[39,7000],[48,5000],[48,6500]])
```

```python
base_salario  #formato de matriz
```

```python
#padronização
scaler_salario = StandardScaler()
base_salario = scaler_salario.fit_transform(base_salario)
```

```python
base_salario #temos que deixar os valores na mesma escala, para o algoritimo nao entender que um atributo é mais importante do que o outro
```

```python
#treinamento
Kmeans_salario = KMeans(n_clusters=3)
Kmeans_salario.fit(base_salario)
```

```python
centroides = Kmeans_salario.cluster_centers_ # centroide é a média de todos os elementos dos grupos
centroides #1 centroide para cada um dos clusters
```

```python
scaler_salario.inverse_transform(Kmeans_salario.cluster_centers_)  #3grupos
```

```python
rotulos = Kmeans_salario.labels_
rotulos  # 3grupos(de 0 a 2)
```

```python
novo_grafico = px.scatter(x = base_salario[:,0], y =base_salario[:,1], color = rotulos) # 0 é pra idade e 1 é salario
novo_grafico_2 = px.scatter(x = centroides[:,0], y = centroides[:,1], size = [12,12,12])
novo_grafico_3 = go.Figure(data=novo_grafico.data + novo_grafico_2.data) #combinar graficos
novo_grafico_3.show()  #um centroide para cada grupo
```

```python
from sklearn.datasets import make_blobs # geração de dados aleatorios
```

```python
x_random, y_random = make_blobs(n_samples=200, centers=5) # centers indica quant clusters quremos
```

```python
x_random, y_random #y 0 a 4 pois temos 5 clusters 0,1,2,3 e 4
```

```python
Kmeans_blobs = KMeans(n_clusters=5)
Kmeans_blobs.fit(x_random)
```

```python
rotl = Kmeans_blobs.predict(x_random)
rotl
```

```python
centroides_ = Kmeans_blobs.cluster_centers_
centroides_
```

```python
graph1 = px.scatter(x = x_random[:,0], y = x_random[:,1], color=rotl)
graph2 = px.scatter(x = centroides_[:,0], y = centroides_[:,1], size=(5,5,5,5,5,))
graph_all = go.Figure(data = graph1.data + graph2.data)
graph_all.show()
```

base real agora  vamos saber se a pessoa vai pagar o cartao de credito baseado no agrupamento dos 6 meses anteriores

```python
base_cartao = pd.read_csv('/content/credit_card_clients.csv', header=1)
```

```python
base_cartao.head()
```

```python
base_cartao.describe()
```

```python
base_cartao['Bill_Total'] = base_cartao['BILL_AMT1'] + base_cartao['BILL_AMT2'] + base_cartao['BILL_AMT3'] + base_cartao['BILL_AMT4'] + base_cartao['BILL_AMT5'] + base_cartao['BILL_AMT6']
```

```python
base_cartao  # AGORA TEMOS O TOTAL, GRACAS AO AGRUPAMENTO
```

```python
base_cartao.head()
```

```python
x_cartao = base_cartao.iloc[:, [1, 25]].values
x_cartao  #limite e  #divida dos clientes em matriz
```

```python
scaler_cartao = StandardScaler()
x_cartao = scaler_cartao.fit_transform(x_cartao)
```

```python
x_cartao  # algoritimo nao ve nenhum dado com mais importancia agora
```

```python
#aqui nao sabemos o numero de cluster ainda, entao vamos descobrir e depois treinamento

wcss = []
for i in range(1,11): #nao podemos gerar o agrupamento com 0 cluster
  print(i)
  Kmeans_cartao = KMeans(n_clusters=i, random_state=0) #randomstate=0 para sempre termos  o mesmo resultado
  Kmeans_cartao.fit(x_cartao)
  wcss.append(Kmeans_cartao.inertia_)
```

```python
wcss  #10 clusters para decidirmos qual usaremos
```

```python
gráfico = px.line(x = range(1,11), y = wcss) # decidir qual cluster sera usado
gráfico.show()
```

pelo gráfico vimos que 10, 9, 8, 7 e 6 clusters a queda é muito grande de um para outro, já de 5 cluster para 4 já menor, então devemos escolher 4 ou 5 cluster, já os clusters 3, 2 e 1 não quase nenhum mudança de tão pouca diferença.

```python
Kmeans_cartao = KMeans(n_clusters=4, random_state=0)
rotúlos = Kmeans_cartao.fit_predict(x_cartao)
```

```python
grafics = px.scatter(x = x_cartao[:,0], y = x_cartao[:,1], color=rotúlos)
grafics.show()
```

eixo x representa limite de cartao de credito(quanto mais pra direita, maior o limite) e y representa os gastos, que represntam o Bill Total

o grupo 0(azul) tem um limite baixo mas tbm gasta pouco, então nao deve ser interessante o banco aumentar o limite desse grupo, ja que nao devem usar.

o grupo 1 (rosa) tem limite de cartao de crédito alto porém não gastam tanto. Nao deve ser interessante pro banco aumentar seus limites, pois ja devem ter um bom limite

o grupo 2(amarelo) é um grupo que gastam mais e nao tem um limite muito alto, deve ser interessante pro banco aumentar o limite para as pessoas de grupo que estão mais no topo no gráfico que gastam mais.

o grupo 3 (vermelho) na parte da direita sao pessoas que tem um limite alto e tbm gastam bastante, deve ser iteressante o banco oferecer mais crédito para essa parte do grupo.

No final, lá em cima temos um OUTLIER, que é a pessoa que mais gasta e que mais tem limite no banco

```python
#aqui podemos descrever esses resultados em uma lista
lista_clientes = np.column_stack((base_cartao , rotúlos))
lista_clientes  # onde vemos os 4 clusters 0 1 2 3 e ja sabemos o que eles representam
```

```python
lista_clientes =  lista_clientes[lista_clientes[:,26].argsort()]
lista_clientes  #deixando alinhado os grupos...veja de 0...... a 3.... ultimo indice
```

aqui está !!! em um cenário real para o banco, bastaria substituir o primeiro indice pelo nome da pessoa, para assim, o banco aumentar ou não o limite do cartão.
