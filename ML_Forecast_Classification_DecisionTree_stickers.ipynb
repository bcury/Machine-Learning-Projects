```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl
```

```python
base = pd.read_csv('/content/GSPC.csv')
```

```python
base.head()
```

não precisamos da coluna Date, então podemos eliminar ela

```python
base.drop(columns=['Date'], axis =1, inplace = True)
```

```python
base.head()
```

```python
base[-2:] # dois ultimos
```

```python
tomorrow = base[-1::]
```

```python
tomorrow
```

```python
dados = base.drop(base[-1::].index, axis =0)
```

```python
dados.tail()
```

```python
dados['target'] = dados['Close'][1:len(dados)].reset_index(drop=True) #criando nosso target
```

```python
dados.tail()
```

```python
previsao = dados[-1::].drop('target', axis =1)
```

```python
previsao #aqui precisamos tirar o target
```

```python
treino = dados.drop(dados[-1::].index, axis =0)
```

```python
treino
```

```python
treino.tail()
```

```python
treino.loc[treino['target'] > treino['Close'],'target'] =1
```

```python
treino.tail()
```

```python
treino.loc[treino['target'] !=1,'target'] =0
```

```python
treino.tail()
```

entao agora temos o target com 0 e 1

```python
#treinando
```

```python
y = treino['target']
x = treino.drop('target', axis =1)
```

```python
x, y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
# algoritimo de arvore de decisão
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
arvore = ExtraTreesClassifier(n_estimators=50, criterion='entropy')
```

```python
arvore.fit(x_train, y_train)
```

```python
forecast = arvore.predict(x_test)
forecast
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, forecast)
```

o resultado da acurácia é ruim, pois precisamos alimentar mais essa base de dados, porém usando serie temporal, poderiamos ter obtido um resultado melhor

Mas podemos usar este modelo como **Previsão**

ou seja, para PREVER as ações futuras

```python
previsao
```

```python
arvore.predict(previsao)
```

```python
previsao['target'] = arvore.predict(previsao)
previsao
```

entao temos que o modelo diz que o fechamento vai ser maior, com o target sendo 1

para ver o que realmente aconteceu na real, temos a var tomorrow para comparar

```python
tomorrow
```

o que confirma a previsão, uhuu!

agora podemos pegar o 'tomorrow' que ja aconteceu e jogarmos na nossa base para treinar o modelo novamente


```python
dados = dados.append(tomorrow, sort=True)
```

```python
dados #temos nossa base de volta
```
